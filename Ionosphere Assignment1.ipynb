```python
 x_train = train.drop(columns=['label'])
```

```python
x_test = test.drop(columns=['label'])
```

```python
y_test = test['label']
```

```python
y_train = pd.Series(map(lambda x: dict(g=1, b=0)[x],
              y_train.values.tolist()), y_train.index)
```

```python
y_test=pd.Series(map(lambda x: dict(g=1, b=0)[x],
              y_test.values.tolist()), y_test.index)
```

```python
from keras import layers
from keras import models
def build_model():
    model = models.Sequential([
        layers.Dense(16, activation='tanh', input_shape = (x_train.shape[1],)),
        layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
    return model
```

```python
model = build_model()
```

```python
model.fit(x_train, y_train, epochs=100)
```

```python
model.evaluate(x_test, y_test, verbose=1)
```

```python
prediction = model.predict(x_test)
```
